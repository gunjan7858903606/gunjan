```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
# from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/content/Linear Regression - Sheet1.csv")
df
```

```python
sns.scatterplot(data=df,x="X",y="Y")
```

```python
x=df.drop(columns=["Y"])
y=df["Y"]
```

```python
x.shape,y.shape
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=24)

# Defining the model
lr = LinearRegression()

# Fitting the model
lr.fit(x_train,y_train)

# Getting predictions on X_test
y_pred = lr.predict(x_test)
# print(y_pred)
```

```python
x_train.head()
```

```python
d={'matric':[],'value':[]}
error_df=pd.DataFrame(data=d)
error_df
```

```python
from sklearn.metrics import mean_absolute_error
d = {"metric": ["MAE"],
     "value": [mean_absolute_error(y_test,y_pred)]}
temp_df = pd.DataFrame(data = d)
error_df = pd.concat([error_df,temp_df],ignore_index = True)
error_df
```

```python
from sklearn.metrics import mean_squared_error

d = {'metric': ["MSE"],
     'value': [mean_squared_error(y_test,y_pred)]}
temp_df = pd.DataFrame(data = d)
error_df = pd.concat([error_df,temp_df],ignore_index = True)
error_df
```

```python
x = np.random.rand(200, 1) + 0.3
```

```python
x.shape
```

```python
x.ndim
```

```python
y = 15*x**2 - 20*x + 19 + (np.random.randn(200, 1))*0.5
y.shape
```

```python
poly_df = pd.DataFrame()
poly_df["X"] = x.flatten()
poly_df["y"] = y.flatten()
poly_df.head()
```

```python
plt.scatter(poly_df["X"], poly_df["y"])
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
x=poly_df.drop(columns=["y"])
y=poly_df["y"]
x.shape,y.shape
```

```python
# Train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=24)

print(f'Shape of x_train -: {x_train.shape}')
print(f'Shape of x_valid -: {x_test.shape}')
print(f'Shape of y_train -: {y_train.shape}')
print(f'Shape of y_valid -: {y_test.shape}')
```

```python
# Defining the model
lr = LinearRegression()

# Fitting the model
lr.fit(x_train,y_train)

# Getting predictions on X_test
y_pred = lr.predict(x_test)
```

```python
# Calculating the r2 score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
plt.plot(x_train, lr.predict(x_train), c ="red")
plt.scatter(x_train,y_train, c = "blue")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(x_train, lr.predict(x_train), c ="red")
# plt.plot(x_test, lr.predict(x_test), c ="green")
plt.scatter(x_train,y_train, c = "blue")
plt.scatter(x_test,y_test, c = "green")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Polynomial Linear Regression of degree 2
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
x_train.head()
```

```python
pd.DataFrame(x_train_poly).head()
```

```python
# Defining the model
poly_lr = LinearRegression()

# Fitting the model
poly_lr.fit(x_train_poly,y_train)

# Getting predictions on X_test
y_pred_poly = poly_lr.predict(x_test_poly)
```

```python
# Calculating the r2 score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_poly)
```

```python
plt.plot(x_train, poly_lr.predict(x_train_poly), c ="red")
# plt.plot(x_test, lr.predict(x_test), c ="green")
plt.scatter(x_train,y_train, c = "blue")
plt.scatter(x_test,y_test, c = "green")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
d = {"x_train":x_train["X"].reset_index(drop=True), "y_train_pred_poly" : pd.Series(poly_lr.predict(x_train_poly))}
temp_df = pd.DataFrame(data = d)

sns.lineplot(data=temp_df, x="x_train", y="y_train_pred_poly", color = "red", label = "Prediction")
plt.scatter(x_train,y_train, c = "blue", label="train_data")
plt.scatter(x_test,y_test, c = "green", label="test_data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
